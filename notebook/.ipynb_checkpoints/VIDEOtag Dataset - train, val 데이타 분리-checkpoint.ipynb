# Make Train Image set, Validation Image set

annotation이 없는 image는 사용하지 않음.
image id를 int로 변경 (string이면 coco.py에서 error 발생함)

```python
used_vt_img_ary = []

for img in vt_img_ary:
    annos = get_annos(img['id'])
    if len(annos) > 0:
        img['id'] = int(img['id'])
        used_vt_img_ary.append(img)

        
total = len(used_vt_img_ary)

val_imgs = used_vt_img_ary[0:50]
train_imgs = used_vt_img_ary[50:total]

print('total', total)
```

# Copy Images

```python
import shutil
from shutil import copyfile

def copy_image(old_path, new_path):
    if not os.path.exists(new_path):
        if not os.path.exists(old_path):
            return 0
        else:
            copyfile(old_path, new_path)
            return 1
    else:
        return 0
    

for img_data in train_imgs:
    img_name = img_data['file_name']
    img_path = os.path.join(args_input_imgs_dir, img_name)
    new_img_path = os.path.join(args_output_train_img_dir, img_name)
    copy_image(img_path, new_img_path)
            
for img_data in val_imgs:
    img_name = img_data['file_name']
    img_path = os.path.join(args_input_imgs_dir, img_name)
    new_img_path = os.path.join(args_output_val_img_dir, img_name)
    copy_image(img_path, new_img_path)
```

# Make Train Annotation set, Validation Annotation set

```python
val_annos = []
train_annos = []

def correction_bbox(bbox, width, height):
    x1 = max(0, bbox[0])
    y1 = max(0, bbox[1])
    w = min(width-x1, bbox[2])
    h = min(height-y1, bbox[3])
    return [x1,y1,w,h]

for img in val_imgs:
    annos = get_annos(img['id'])
    for anno in annos:
        anno['bbox'] = correction_bbox(anno['bbox'], 640, 360)
        val_annos.append(anno)
    
for img in train_imgs:
    annos = get_annos(img['id'])
    for anno in annos:
        anno['bbox'] = correction_bbox(anno['bbox'], 640, 360)
        train_annos.append(anno)
    
print(len(val_annos))
print(len(train_annos))
```
